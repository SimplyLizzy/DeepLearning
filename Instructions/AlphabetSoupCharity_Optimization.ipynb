## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

from keras.models import Model

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("../Resources/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID column, 'EIN'.
app_df = application_df.drop('EIN', axis = 1)
app_df
```

```python
# Determine the number of unique values in each column.
app_df.nunique()
```

```python
# Look at NAME value counts for binning
app_df['NAME'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `name_types_to_replace`
name_types_to_replace = app_df['NAME'].value_counts()
name_types_to_replace = name_types_to_replace[name_types_to_replace<10].index
name_types_to_replace

# Replace in dataframe
for app in name_types_to_replace:
    app_df['NAME'] = app_df['NAME'].replace(app,"Other")

# Check to make sure binning was successful
app_df['NAME'].value_counts()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
application_test = app_df['APPLICATION_TYPE'].value_counts()
application_test = application_test[application_test < 500].index
application_test

# Replace in dataframe
for cls in application_test:
    app_df['APPLICATION_TYPE'] = app_df['APPLICATION_TYPE'].replace(cls,"Other")

# Check to make sure binning was successful
app_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
app_df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_test = app_df['CLASSIFICATION'].value_counts()
classification_test = classification_test[classification_test > 1]
classification_test
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = app_df['CLASSIFICATION'].value_counts()
classifications_to_replace = classifications_to_replace[classifications_to_replace < 1800].index
classifications_to_replace

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
app_dummies = pd.get_dummies(app_df)
print(app_dummies.columns)
app_dummies
```

```python
# Split our preprocessed data into our features and target arrays
features = app_dummies.drop('IS_SUCCESSFUL', axis = 1).values
target = app_dummies['IS_SUCCESSFUL'].values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=47)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
input = len(X_train[0])

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=15, activation="sigmoid", input_dim=input))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=10, activation="sigmoid"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# Create a callback that saves the model's weights every 5 epochs.
callback = tf.keras.callbacks.ModelCheckpoint('weights/weights{epoch:08d}.h5', save_weights_only=True, verbose = 0)
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save("AlphabetSoupCharity_Optimization.h5")
```
